```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
train_data_1 = pd.read_csv('/content/gdrive/MyDrive/celebal3/Training Data/Diabetes_XTrain.csv')
train_data_2 = pd.read_csv('/content/gdrive/MyDrive/celebal3/Training Data/Diabetes_YTrain.csv')
```

```python
test_data = pd.read_csv('/content/gdrive/MyDrive/celebal3/Testing Data/Diabetes_Xtest.csv')
```

```python
x=train_data_1.values
y=train_data_2.values
y.reshape((-1))
print(x.shape)
print(y.shape)
n= train_data_1.nunique()
n
```

```python
label_x = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
label_y = []
i=0
for i in range(8):
    label_y.append(n[i])
plt.scatter(label_x,label_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x,y)
```

```python
test_data = np.asarray(test_data)
```

```python
model.predict(test_data)
```

```python
model.score(x,y)
```
